```python
import numpy as np
import matplotlib.pyplot as plt

#     
X = np.random.randint(-20, 21, size = 30)
Y = np.random.randint(-20, 21, size = 30)

#  
plt.plot(X, Y, 'g') 
#  
plt.show() 
```

# Bubble sort

```python
#   
for i in range(0, len(X) - 1):  
    for j in range(0, len(X) - i - 1):
        if X[j] > X[j+1]:
            X[j], X[j+1] = X[j+1], X[j]
            
plt.plot(X,Y, 'r.')
plt.show()
```
