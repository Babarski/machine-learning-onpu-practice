```python
import graphlab
data = graphlab.SFrame('Mass-Shootings-Dataset.csv')
```

```python
data.head()
```

1) Рассчитать TF-IDF для столбца Summary.

```python
tfidf = graphlab.text_analytics.tf_idf(data['Summary'])
tfidf
```

2) Заменить в столбце Mental Health Issues значения на [0, 1, 2], где:
    0 - No
    1 - Yes
    2 - Unclear/Unknown и другие значения

```python
def change_mental(s):
    if s == "Yes":
        return 1
    elif s == "No":
        return 0
    else:
        return 2
    
data['Mental Health Issues'] = data['Mental Health Issues'].apply(change_mental)
data.head()
```

3) На основе примера обучить модель с использованием такого набора фич:
    [‘tf_idf’, ‘Total victims’, ’Mental Health Issues’] для

        3 кластеров

```python
import matplotlib.pyplot as plt
%matplotlib inline
# , data['tfidf'], '.g'
plt.plot(data['Total victims'], '.r', data['Mental Health Issues'], '.b')
```

```python
my_features = ['Total victims', 'Mental Health Issues', 'tfidf']
model = graphlab.kmeans.create(data,
                               features = my_features,
                               num_clusters=3)
```

```python
model['cluster_id'].head()
```

```python
graphlab.canvas.set_target('ipynb')
model.show()
```

```python
data['cluster'] = model['cluster_id']['cluster_id']
data
```

```python
data[data['cluster'] == 1]
```

```python
data[data['cluster'] == 0]
```

```python
data[data['cluster'] == 2]
```
