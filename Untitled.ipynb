```python
import numpy as np
import matplotlib.pyplot as plt

# создание массивов со случайными числами
X = 40*np.random.random_sample(30) -20 
Y = 40*np.random.random_sample(30) -20

# создание графика
plt.plot(X, 'g') 
# демонстрация графика
plt.show() 
```

# Bubble sort

```python
# реализация сортировки пузырьком
for i in range(0, len(X) - 1):  
    for j in range(0, len(X) - i - 1):
        if X[j] > X[j+1]:
            X[j], X[j+1] = X[j+1], X[j]
            
plt.plot(X, 'r.')
plt.show()
```

```python
for i in range(0, len(Y)-1):
    min = i
    for j in range (i+1, len(Y)):
        if Y[j]< Y[min]:
            min = j
    if min != i:
        Y[min],Y[i] = Y[i],Y[min]

plt.plot(Y, 'r.')
plt.show()
```
